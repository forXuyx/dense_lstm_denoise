```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
from matplotlib.colors import LinearSegmentedColormap
import matplotlib
from matplotlib import gridspec
from matplotlib.ticker import FuncFormatter
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

# 散点图

```python
cmap = plt.cm.Spectral

# 随机生成一些例子，可以替换为你自己的
target_snr_e5 = np.random.uniform(5, 55, 1000)
denoised_snr_e5 = target_snr_e5 + np.random.uniform(-1, 1, 1000)
overlap_all_e5 = np.random.uniform(0.9, 1, 1000)

target_snr_e6 = np.random.uniform(5, 55, 1000)
denoised_snr_e6 = target_snr_e6 + np.random.uniform(-1, 1, 1000)
overlap_all_e6 = np.random.uniform(0.9, 1, 1000)

target_snr_e7 = np.random.uniform(5, 55, 1000)
denoised_snr_e7 = target_snr_e7 + np.random.uniform(-1, 1, 1000)
overlap_all_e7 = np.random.uniform(0.9, 1, 1000)

with plt.style.context(['science', 'scatter', 'nature']):
    fig, axs = plt.subplots(1, 3, figsize=(13, 4), constrained_layout=True)
    
    sc1 = axs[0].scatter(target_snr_e5, denoised_snr_e5, c=overlap_all_e5, cmap=cmap, vmin=0.9, vmax=1, s=50)
    axs[0].set_title(r'$Mc=1\times10^5$', fontsize=16)
    axs[0].plot([5, 55], [5, 55], 'w--')
    axs[0].set_xlim(5, 55)
    axs[0].set_ylim(5, 55)
    axs[0].grid(True)
    axs[0].tick_params(axis='both', labelsize=16)

    sc2 = axs[1].scatter(target_snr_e6, denoised_snr_e6, c=overlap_all_e6, cmap=cmap, vmin=0.9, vmax=1, s=50)
    axs[1].set_title(r'$Mc=1\times10^6$', fontsize=16)
    axs[1].plot([5, 55], [5, 55], 'w--')
    axs[1].set_xlim(5, 55)
    axs[1].set_ylim(5, 55)
    axs[1].grid(True)
    axs[1].tick_params(axis='both', labelsize=16)

    sc3 = axs[2].scatter(target_snr_e7, denoised_snr_e7, c=overlap_all_e7, cmap=cmap, vmin=0.9, vmax=1, s=50)
    axs[2].set_title(r'$Mc=1\times10^7$', fontsize=16)
    axs[2].plot([5, 55], [5, 55], 'w--')
    axs[2].set_xlim(5, 55)
    axs[2].set_ylim(5, 55)
    axs[2].grid(True)
    axs[2].tick_params(axis='both', labelsize=16)

    # 添加共用的颜色条
    cbar = fig.colorbar(sc1, ax=axs, orientation='vertical', label='Overlap')
    cbar.ax.set_ylabel('Overlap', fontsize=16)

    # 设置共用的x轴和y轴标签
    fig.text(0.5, -0.015, r'$SNR_{target}$', ha='center', va='center', fontsize=16)
    fig.text(-0.01, 0.5, r'$SNR_{denoised}$', ha='center', va='center', rotation='vertical', fontsize=16)

    # 保存图片
    plt.savefig('test1.pdf', format='pdf')

    plt.show()
```

# 柱状图

```python
cmap = plt.cm.Spectral

# 随机数据
overlap_all_e5 = np.random.uniform(0.99, 1, 1000)
overlap_all_e6 = np.random.uniform(0.99, 1, 1000)
overlap_all_e7 = np.random.uniform(0.99, 1, 1000)
snr_loss_e5 = np.random.uniform(15, 20, 1000)
snr_loss_e6 = np.random.uniform(15, 20, 1000)
snr_loss_e7 = np.random.uniform(15, 20, 1000)

with plt.style.context(['science', 'nature']):
    fig, axs = plt.subplots(1, 2, figsize=(10, 4), constrained_layout=True)
    
    # Plot histograms with only the outline
    bins = np.linspace(0.9, 1, 50)
    axs[0].hist(overlap_all_e5, bins=bins, color='b', histtype='step', linewidth=2, label=r'$Mc=1\times10^5$', alpha=0.5)
    axs[0].hist(overlap_all_e6, bins=bins, color='r', histtype='step', linewidth=2, label=r'$Mc=1\times10^6$', alpha=0.5)
    axs[0].hist(overlap_all_e7, bins=bins, color='g', histtype='step', linewidth=2, label=r'$Mc=1\times10^7$', alpha=0.5)
    axs[0].tick_params(axis='both', labelsize=16)
    axs[0].set_xlabel('Overlap', fontsize=16)
    axs[0].set_ylabel('Count', fontsize=16)
    axs[0].legend(fontsize=16)

    bins = np.linspace(0, 20, 50)
    axs[1].hist(snr_loss_e5, bins=bins, color='b', histtype='step', linewidth=2, label=r'$Mc=1\times10^5$', alpha=0.5)
    axs[1].hist(snr_loss_e6, bins=bins, color='r', histtype='step', linewidth=2, label=r'$Mc=1\times10^6$', alpha=0.5)
    axs[1].hist(snr_loss_e7, bins=bins, color='g', histtype='step', linewidth=2, label=r'$Mc=1\times10^7$', alpha=0.5)
    axs[1].tick_params(axis='both', labelsize=16)
    axs[1].set_xlabel('SNR Loss (\%)', fontsize=16)
    axs[1].legend(fontsize=16)

    plt.tight_layout()

    plt.savefig('test2', format='pdf')

    plt.show()
```

# 降噪图

```python

with plt.style.context(['science', 'notebook']):
    fig = plt.figure(figsize=(32, 6))
    gs = gridspec.GridSpec(4, 5, figure=fig)

    font_size = 20  # General font size for text
    tick_size = 20  # Size for tick labels


    time = np.linspace(0, 1, 1000)  # Example time array
    target_wave = np.sin(2 * np.pi * 50 * time) * 0.1
    denoised_wave = target_wave + np.random.normal(0, 0.01, target_wave.shape)

    # Adjusting all ax1 and ax2 configurations consistently
    for i in range(3):
        title = [r'$Mc=1\times10^5$', r'$Mc=1\times10^6$', r'$Mc=1\times10^7$'][i]
        
        ax1 = fig.add_subplot(gs[:3, i])
        ax1.plot(time, target_wave, label='Target wave', linewidth=2, color='blue')
        ax1.plot(time, denoised_wave, '--', label='Denoised wave', linewidth=2, color='orange')
        if i == 0:
            ax1.set_ylabel(r"Whitened strain", fontsize=font_size)
        ax1.legend(loc='upper left', fontsize=tick_size)
        info_text = 'Target SNR={:.4f}\nDenoised SNR={:.4f}\nOverlap={:.4f}'.format(20, 21, 0.9)
        ax1.text(0.05, 0.05, info_text, transform=ax1.transAxes, ha='left', va='bottom', fontsize=tick_size)
        ax1.tick_params(labelbottom=False, labelsize=tick_size)
        ax1.ticklabel_format(style='scientific', axis='y', scilimits=(0,0))
        ax1.set_title(title, fontsize=font_size)

        ax2 = fig.add_subplot(gs[3:, i], sharex=ax1)
        residual = denoised_wave - target_wave
        ax2.plot(time, residual, linewidth=2, color='red')
        ax2.set_xlabel(r"Time [s]", fontsize=font_size)
        if i == 0:
            ax2.set_ylabel(r"Residual", fontsize=font_size)
        ax2.ticklabel_format(style='scientific', axis='y', scilimits=(0,0))
        ax2.tick_params(labelsize=tick_size)

    plt.tight_layout()
    plt.savefig('test3.pdf', format='pdf')
    plt.show()

```
